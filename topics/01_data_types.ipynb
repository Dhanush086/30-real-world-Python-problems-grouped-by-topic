üíä Pharmacy Billing System

A pharmacist wants to bill a customer for their medicine purchase. The details include:

Customer name (string)

Medicine name (string)

Quantity bought (int)

Price per tablet (float)

Is prescription required (bool)

Write a Python program using these data types and display the bill.

medical billing.jpg

```python
# --- Pharmacy Billing System ---

# Data (using basic data types)
customer_name = "Rohan Gupta"           # string
medicine_name = "Paracetamol 500mg"     # string
quantity = 20                           # int
price_per_unit = 2.50                   # float
prescription_required = False           # bool

# Display Bill
print("--- Pharmacy Bill ---")
print(f"Customer: {customer_name}")
print(f"Medicine: {medicine_name}")
print(f"Quantity: {quantity}")
print(f"Price per unit: ‚Çπ{price_per_unit}")
print(f"Prescription needed: {prescription_required}")

# Data Type Check
print("\n--- Data Types ---")
print(type(customer_name))
print(type(medicine_name))
print(type(quantity))
print(type(price_per_unit))
print(type(prescription_required))
```
